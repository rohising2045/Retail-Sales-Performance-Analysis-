```python
import numpy as np
import pandas as pd 
```

```python
sales = pd.read_csv('retail_sales_10000.csv')
store = pd.read_csv('stores_50.csv')
merged = pd.read_csv('merged_data.csv')
```

TOTAL REVENUE BY REGION

```python
print(merged.head(5))
```

TOTAL REVENUE ,TOTAL COST & TOTAL REVENUE BY REGION

```python
Total_Revenue = merged['revenue'].sum()
print("Total Revenue: ",Total_Revenue)

Total_cost = merged['cost'].sum()
print("Total cost: ",Total_cost)

revenue_by_region = merged.groupby('region')['revenue'].sum()
print("Revenue by Region: ", revenue_by_region)
```

TOTAL PROFIT & TOTAL PROFIT BY REGION

```python

print("Total_profit: ",merged['profit'].sum())

print("TOTAL PROFIT BY REGION: ", merged.groupby('region')['profit'].sum().sort_values(ascending=False))
```

TOP STORED

```python
top_stores = merged.groupby('store_id')['revenue'].sum()

top_stores = top_stores.sort_values(ascending=False)

print("Top 5 Stores by Revenue: \n", top_stores.head(5))
```

CATEGORY WISE TOTAL UNIT SOLD

```python
unit_by_category = merged.groupby('product_category')['units_sold'].sum()
print(unit_by_category)
```

MONTHLY REVENUE TREND

```python
merged['date'] = pd.to_datetime(merged['date'])
merged['month'] = merged['date'].dt.to_period('M')  
monthly_revenue = merged.groupby('month')['revenue'].sum().reset_index()
monthly_revenue = monthly_revenue.sort_values('month')
print(monthly_revenue)
print("\nHIGHEST REVENUE MONTH: \n",monthly_revenue.max())
```

```python

monthly_trend = merged.groupby('month')[['revenue', 'profit']].sum().reset_index()
print(monthly_trend)

```
