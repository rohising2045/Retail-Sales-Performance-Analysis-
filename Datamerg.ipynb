```python
import numpy as np
import pandas as pd 
```

```python
sales = pd.read_csv('retail_sales_10000.csv')
print(sales.head(10))

store = pd.read_csv('stores_50.csv')
print(store.head(10))
```

MERGING BOTH DATASET

```python
merged = pd.merge(sales, store, on="store_id", how="left")
merged.head()
```

CREATING NEW COLUMNS USING BOTH THE DATASET
1: Total Revenue
2: Total cost 
3: Total Profit


```python
merged['revenue']= merged['units_sold']* merged['unit_price']
merged['cost']= merged['units_sold']*merged['cost_per_unit']
merged['profit']= merged['revenue']-merged['cost']

print(merged.head(5))
```

```python
merged.to_csv('merged_data.csv',index=False)
```
