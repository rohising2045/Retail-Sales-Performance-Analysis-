CREATTING THE PRIMARY DATABASE 

```python
import pandas as pd
import numpy as np

np.random.seed(42)
dates = pd.date_range(start="2023-01-01", periods=10000, freq="H")
data = pd.DataFrame({
    "date": np.random.choice(dates, 10000),
    "transaction_id": np.arange(1, 10001),
    "store_id": np.random.randint(1001, 1051, size=10000),
    "product_category": np.random.choice(["Electronics","Clothing","Groceries","Furniture","Sports"], 10000),
    "units_sold": np.random.randint(1, 20, 10000),
    "unit_price": np.random.uniform(50, 2000, 10000).round(2),
    "cost_per_unit": np.random.uniform(30, 1500, 10000).round(2)
})
data.to_csv("retail_sales_10000.csv", index=False)
print("retail_sales_10000.csv saved successfully")
```

CREATING SECONDRY(STORE) DATASET

```python
import pandas as pd
import numpy as np

np.random.seed(42)
store_ids = np.arange(1001, 1051)
store = pd.DataFrame({
    "store_id": store_ids,
    "store_name": [f"Store_{i}" for i in store_ids],
    "city": np.random.choice(["Delhi","Mumbai","Bangalore","Hyderabad","Kolkata","Pune"], 50),
    "region": np.random.choice(["North","South","East","West"], 50),
    "store_size_sqft": np.random.randint(1000, 5000, 50),
    "opening_date": pd.to_datetime(np.random.choice(pd.date_range("2015-01-01","2023-12-31"), 50))
})
store.to_csv("stores_50.csv", index=False)
print("✅ stores_50.csv saved successfully")
```
